# Getting the Table on the Main Page

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://www.worldometers.info/world-population/population-by-country/'
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req)
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
table = soup.find('table')
links = [link['href'] for link in table.find_all('a', href=True)]

links[0]
```

# Getting Data from the Graph

```python
data_dict={}
```

```python
for i in links:
    url = 'https://www.worldometers.info' + i
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req)
    soup = BeautifulSoup(webpage, 'html.parser')

    script_tag = soup.find_all('script')

    years = re.findall(r'v:\t(.+?)\t}]},', str(script_tag[7]))
    first_year = re.findall('v:\t(.+?) }]},', str(script_tag[7]))
    years = first_year + years

    country = re.findall('-population/(.+?)-population/', i)
    int_years = [int(i) for i in years] 
    data_dict[country[0]] = int_years
data_dict['year'] = list(range(1950, 2020))
data_dict
```

```python
df=pd.DataFrame.from_dict(data_dict,orient='index').transpose()
df.set_index('year', inplace = True)
df.head()
```

```python
df.to_excel('population_country_1950_2019.xls')
```
